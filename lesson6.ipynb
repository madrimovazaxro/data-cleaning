```python
import opendatasets as od
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
od.download("https://www.kaggle.com/competitions/house-prices-advanced-regression-techniques/data")
```

```python
df = pd.read_csv("/content/house-prices-advanced-regression-techniques/train.csv")
print(df.shape)
df.head()
```

```python
df.info()
```

```python
#1.Umumiy soni
df.isnull().sum().sort_values(ascending=False).head(15)

#The code finds the top 15 columns in df that contain the most missing (NaN) values,
#sorted from highest to lowest.
```

```python
#2.Foizda
miss = (df.isnull().sum()/len(df)*100).round(2)
miss[miss>0].sort_values(ascending=False).head(15)

# This code computes the percentage of missing values for each column,
# removes columns with no missing values, sorts the remaining columns by missing percentage
# (highest first), and displays the top 15.
```

```python
#3.Vizual (heatmap)
sns.heatmap(df.isnull(), cbar=True, cmap="Reds")
plt.show()

#cbar -
```

```python
df.drop(['PoolQC', 'MiscFeature', 'Alley', 'Fence', 'MasVnrType', 'FireplaceQu'], axis=1, inplace = True)

#axis=1 -
#inplace= -
```

```python
df.isnull().sum().sort_values(ascending=False).head(15)
```

```python
#Barcha kategorik ustunlarni to'ldirish
cat_all = df.select_dtypes(include='object').columns
df[cat_all] = df[cat_all].fillna('None')
print(df[cat_all])

#Raqamlilarni to'ldirish
num = df.select_dtypes(include = "number").columns
df[num]=df[num].fillna(df[num].median())

#Tekshirish
print(df.isnull().sum().sum())
```

```python
#1.Duplicate tekshirish
print(df.duplicated().sum()) #soni
#df.drop_duplicates(inplace=True) #o'chirish
```

```python
#2.Barcha dtype ko'rish
df.dtypes.value_counts()
```

```python
#3. MSSubClass: raqam -> kategorik
df['MSSubClass'] = df['MSSubClass'].astype(str)
```

```python
#4. Date ustuni borligini tekshirib ko'rish
"Date" in df.columns
#agar bo'lganda
# df['date'] = pd.to_datetime(df['date'])
```

```python
'SalePrice' in df.columns
```

```python
df['SalePrice']
```

```python
#quantile() -
Q1 = df['SalePrice'].quantile(0.25)
print(Q1)
Q3 = df['SalePrice'].quantile(0.75)
print(Q3)
IQR = Q3 - Q1

#Outlier filtr
mask = (df["SalePrice"] >= Q1 - 1.5 * IQR) & \
        (df['SalePrice'] <= Q3 + 1.5 * IQR)

print(mask)
df_clean=df[mask]
```

```python
#1. SalePrice boxplot
fig,axes=plt.subplots(1,2, figsize=(12,6))

sns.boxplot(data=df, y='SalePrice', ax=axes[0])
axes[0].set_title('Outlier bor')
sns.boxplot(data=df_clean, y='SalePrice', ax=axes[1])
axes[1].set_title("Outlier olib tashlangan")
plt.tight_layout()
plt.show()
```

```python
#1. SalePrice boxplot
fig,axes=plt.subplots(1,2, figsize=(12,6))

sns.violinplot(data=df, y='SalePrice', ax=axes[0])
axes[0].set_title('Outlier bor')
sns.violinplot(data=df_clean, y='SalePrice', ax=axes[1])
axes[1].set_title("Outlier olib tashlangan")
plt.tight_layout()
plt.show()
```

```python
#1. SalePrice boxplot
fig,axes=plt.subplots(1,2, figsize=(12,6))

sns.stripplot(data=df, y='SalePrice', ax=axes[0])
axes[0].set_title('Outlier bor')
sns.stripplot(data=df_clean, y='SalePrice', ax=axes[1])
axes[1].set_title("Outlier olib tashlangan")
plt.tight_layout()
plt.show()
```

```python
#1. SalePrice boxplot
fig,axes=plt.subplots(1,2, figsize=(12,6))

sns.swarmplot(data=df, y='SalePrice', ax=axes[0])
axes[0].set_title('Outlier bor')
sns.swarmplot(data=df_clean, y='SalePrice', ax=axes[1])
axes[1].set_title("Outlier olib tashlangan")
plt.tight_layout()
plt.show()
```
